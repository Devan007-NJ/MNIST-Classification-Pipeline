#**Digit Recognition Model**

```python
import pandas as pd
import numpy as np
```

##Loading the MNIST data

```python
from sklearn.datasets import fetch_openml
data,target= fetch_openml('mnist_784',version=1,return_X_y=True,as_frame=False)
data=data.astype(np.float32)/255.0
target=target.astype(int)
```

###Splitting Data to Training and Testing Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data,target,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
y_test
```

##Creating the Model


```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(X_train,y_train)
```

##Testing and Evalutation of Model

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
print("The accuracy is close to 97 percentage")
```

##Visualization of the trained model's performance

```python
import matplotlib.pyplot as plt
import seaborn as sns
confusion_matirx=pd.crosstab(y_test,y_pred,rownames=['Actual'],colnames=['Predicted'])
sns.heatmap(confusion_matirx,annot=True)
plt.show()
```
